```python
import klib
import pandas as pd
```

```python
# Data paths definitions
SLEEP_PATH = '../../data/sleep.csv'
SPORT_PATH = '../../data/sport.csv'
HR_PATH = '../../data/heartrate_auto.csv'
ACTIVITY_PATH = '../../data/activity.csv'
```

```python
df_activity = pd.read_csv(ACTIVITY_PATH)
```

```python
df_activity.head()
```

## Correlation plot for Activity data

```python
klib.corr_mat(df_activity)
```

```python
klib.corr_plot(df_activity[['steps', 'distance', 'run_distance', 'calories']], split='pos')
```

## Selective correlation plot

```python
klib.corr_plot(df_activity, target='calories')
```

## Distribution plots for Activity data

```python
klib.dist_plot(df_activity['distance'])
```

```python
klib.dist_plot(df_activity['run_distance'])
```

```python
klib.dist_plot(df_activity['steps'])
```

```python
klib.dist_plot(df_activity['calories'])
```

```python
# Missing values check
klib.missingval_plot(df_activity)
```
